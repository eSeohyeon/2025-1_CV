```python
# 샘플 RGB 이미지
rgb_img = tf.random.uniform([100, 100, 3], maxval=255, dtype=tf.float32)
print(rgb_img)
```

```python
plt.imshow(rgb_img)
plt.title("RGB Image")
plt.axis('off')
plt.show()
```

```python
grayscale_img = tf.image.rgb_to_grayscale(rgb_img)

print(grayscale_img.shape)
plt.imshow(grayscale_img.numpy().squeeze(), cmap='gray')
plt.title('Grayscale Image')
plt.axis("off")
plt.show()
```

```python
#  (0, 0) 에서의 r, g, b 값에 가중치를 곱함 -> 그레이스케일 값
R = rgb_img[0][0][0] * 0.299
G = rgb_img[0][0][1] * 0.587
B = rgb_img[0][0][2] * 0.114
V = R+G+B

# 실제 값, 계산 값
print(grayscale_img[0][0], V)
```

```python
hsv_img = tf.image.rgb_to_hsv(rgb_img)

hue_channel = hsv_img[:,:,0]

plt.imshow(hue_channel, cmap='hsv')
plt.title('Hue Channel of HSV Image')
plt.axis('off')
plt.colorbar(label='Hue Value')
plt.show()
```

```python
# 픽셀값의 정규화와 표준화

# 정규화
normalized_img = rgb_img / 255.0
rgb_img[0][0], normalized_img[0][0]
```

```python
# 표준화
mean = tf.reduce_mean(rgb_img)
stddev = tf.math.reduce_std(rgb_img)

standardized_img = (rgb_img - mean) / stddev
rgb_img[0][0], standardized_img[0][0]
```

```python
# 중앙 값 필터링

import numpy as np
import cv2

def generate_salt_noise(image):
  num_salt = np.ceil(0.05 * image.size)
  coords = [np.random.randint(0, i-1, int(num_salt)) for i in image.shape]
  salted_image = image.copy()
  salted_image[coords[0], coords[1]] = 255
  return salted_image

def generate_pepper_noise(image):
  num_pepper = np.ceil(0.05 * image.size)
  coords = [np.random.randint(0, i-1, int(num_pepper)) for i in image.shape]
  peppered_image = image.copy()
  peppered_image[coords[0], coords[1]] = 0
  return peppered_image
```

```python
!wget https://raw.githubusercontent.com/Cobslab/imageBible/main/image/like_lenna.png
lenna_img = cv2.imread('like_lenna.png', cv2.IMREAD_GRAYSCALE)

salted_lenna = generate_salt_noise(lenna_img)
peppered_lenna = generate_pepper_noise(salted_lenna)
filtered_lenna = cv2.medianBlur(peppered_lenna, 5)

fig, axes = plt.subplots(1, 4, figsize = (20, 6))

axes[0].imshow(lenna_img, cmap='gray')
axes[0].set_title('Original Lenna Image')
axes[0].axis('off')

axes[1].imshow(salted_lenna, cmap='gray')
axes[1].set_title('Salted Lenna Image')
axes[1].axis('off')

axes[2].imshow(peppered_lenna, cmap='gray')
axes[2].set_title('Salted & Peppered Lenna')
axes[2].axis('off')

axes[3].imshow(filtered_lenna, cmap='gray')
axes[3].set_title('Median Filtered Lenna')
axes[3].axis('off')

plt.tight_layout()
plt.show()
```

```python
# 가우시안 필터링

img = cv2.imread('like_lenna.png', cv2.IMREAD_GRAYSCALE)
mean = 0
sigma = 1
gaussian_noise = np.random.normal(mean, sigma, img.shape).astype('uint8')
noisy_img = cv2.add(img, gaussian_noise)

plt.imshow(noisy_img, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')
plt.show()
```

```python
sigma_values = [1, 5, 10]
denoised_images = []

for sigma in sigma_values:
  denoised = cv2.GaussianBlur(noisy_img, (0, 0), sigma)
  denoised_images.append(denoised)

fig, axes = plt.subplots(1, 4, figsize = (20, 10))

axes[0].imshow(noisy_img, cmap='gray')
axes[0].set_title('Noisy Image')
axes[0].axis('off')

for ax, img, sigma in zip(axes[1:], denoised_images, sigma_values):
  ax.imshow(img, cmap='gray')
  ax.set_title(f"Denoised (σ={sigma})")
  ax.axis('off')
```

```python
# 아핀 변환

image_path = "like_lenna.png"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
# 회전 변환

def rotate_image(image, angle, center=None):
  rows, cols, _ = image.shape
  if center is None:
    center = (cols // 2, rows // 2)
  M = cv2.getRotationMatrix2D(center, angle, 1)
  rotated = cv2.warpAffine(image, M, (cols, rows))
  return rotated

rotated_img = rotate_image(img, 45)
plt.imshow(rotated_img)
plt.title("Rotated Image")
plt.show()
```
